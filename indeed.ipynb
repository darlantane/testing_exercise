```python
%%writefile indeed_tests.robot

*** Settings ***
Library    SeleniumLibrary
Suite Setup    Open Browser To Indeed
Suite Teardown    Close Browser
Test Setup    Go To Homepage

*** Variables ***
${URL}            https://fr.indeed.com
${BROWSER}        Chrome
${SEARCH_TERM}    informatique
${LOCATION}       Paris
${TIMEOUT}        15 seconds

*** Keywords ***
Open Browser To Indeed
    Open Browser    ${URL}    ${BROWSER}
    Maximize Browser Window
    Set Selenium Timeout    ${TIMEOUT}

Go To Homepage
    Go To    ${URL}
    Accept Cookies

Accept Cookies
    Run Keyword And Ignore Error
    ...    Wait Until Element Is Visible    id=onetrust-accept-btn-handler    5 seconds
    Run Keyword And Ignore Error
    ...    Click Button    id=onetrust-accept-btn-handler

Launch Job Search
    Wait Until Element Is Visible    name=q
    Input Text    name=q    ${SEARCH_TERM}
    Input Text    name=l    ${LOCATION}
    Press Keys    name=l    RETURN
    Wait Until Page Contains Element    css=div.job_seen_beacon    ${TIMEOUT}

Click First Job
    Wait Until Element Is Visible    css=h2.jobTitle
    Click Element    xpath=(//h2[contains(@class,"jobTitle")]//a)[1]
    Wait Until Page Contains Element    css=h1

Go To Next Page
    Wait Until Element Is Visible    css=a[aria-label="Suivant"]
    Click Element    css=a[aria-label="Suivant"]
    Wait Until Page Contains Element    css=div.job_seen_beacon

Go To Companies Page
    Click Link    xpath=//a[contains(@href,"cmp")]
    Wait Until Page Contains Element    tag=body

*** Test Cases ***
Homepage Should Load
    Page Should Contain Element    tag=body

Page Title Should Contain Indeed
    Title Should Contain    Indeed

Search Fields Should Be Visible
    Element Should Be Visible    name=q
    Element Should Be Visible    name=l

User Can Search For Jobs
    Launch Job Search
    Page Should Contain Element    css=div.job_seen_beacon

First Result Should Be Clickable
    Launch Job Search
    Click First Job

User Can Navigate To Next Page
    Launch Job Search
    Go To Next Page

User Can Access Companies Page
    Go To Companies Page
```
